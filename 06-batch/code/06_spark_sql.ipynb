```python
import pyspark
import pyspark.sql.functions as F
from pyspark.sql import SparkSession
```

```python
spark = SparkSession.builder \
    .appName("taxi-data") \
    .getOrCreate()
```

```python
green_df = spark.read.parquet("../data/parquet_spark/green/*/*/*")
```

```python
green_df.count()
```

```python
green_df.columns
```

```python
green_df = green_df\
    .withColumnsRenamed({
        "lpep_pickup_datetime": "pickup_datetime",
        "lpep_dropoff_datetime": "dropoff_datetime",
    }) \
    .withColumn("service_type", F.lit("green"))

green_df.columns
```

```python
green_df.show(5)
```

```python
yellow_df = spark.read.parquet("../data/parquet_spark/yellow/*/*/*")
```

```python
yellow_df.count()
```

```python
yellow_df.columns
```

```python
yellow_df = yellow_df \
    . withColumnsRenamed({
        "tpep_pickup_datetime": "pickup_datetime",
        "tpep_dropoff_datetime": "dropoff_datetime",
    }) \
    .withColumn("service_type", F.lit("yellow"))
yellow_df.columns
```

```python
yellow_df.show(5)
```

```python
# Find the intersection of the two dataframes
# This method doesn't perserve order
set(green_df.columns) & set(yellow_df.columns)
```

```python
# Perserve Green column order
columns = []

for column in green_df.columns:
    if column in yellow_df.columns:
        columns.append(column)

columns
```

```python
df_taxi_data = green_df.select(columns).union(yellow_df.select(columns))
```

```python
df_taxi_data.groupBy("service_type").count().show()
```

```python
df_taxi_data.createOrReplaceTempView("taxi_data")
```

```python
df_result = spark.sql("""
    SELECT 
        -- Reveneue grouping 
        PULocationID AS revenue_zone,
        date_trunc('month', pickup_datetime) AS revenue_month, 
        service_type, 

        -- Revenue calculation 
        SUM(fare_amount) AS revenue_monthly_fare,
        SUM(extra) AS revenue_monthly_extra,
        SUM(mta_tax) AS revenue_monthly_mta_tax,
        SUM(tip_amount) AS revenue_monthly_tip_amount,
        SUM(tolls_amount) AS revenue_monthly_tolls_amount,
        SUM(improvement_surcharge) AS revenue_monthly_improvement_surcharge,
        SUM(total_amount) AS revenue_monthly_total_amount,
        SUM(congestion_surcharge) AS revenue_monthly_congestion_surcharge,

        -- Additional calculations
        AVG(passenger_count) AS avg_monthly_passenger_count,
        AVG(trip_distance) AS avg_monthly_trip_distance
    FROM
        taxi_data
    GROUP BY
        1, 2, 3
""")
df_result.show()


```

```python
df_result \
    .coalesce(1) \
    .write.parquet("../data/parquet_spark/revenue", mode="overwrite")
```
