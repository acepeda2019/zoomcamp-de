```python
green_schema = types.StructType([
    types.StructField('VendorID', types.IntegerType(), True),                          # int64
    types.StructField('lpep_pickup_datetime', types.TimestampType(), True),     # datetime64[us]
    types.StructField('lpep_dropoff_datetime', types.TimestampType(), True),    # datetime64[us]
    types.StructField('store_and_fwd_flag', types.StringType(), True),                  # str
    types.StructField('RatecodeID', types.DoubleType(), True),                      # float64
    types.StructField('PULocationID', types.IntegerType(), True),                      # int64
    types.StructField('DOLocationID', types.IntegerType(), True),                      # int64
    types.StructField('passenger_count', types.DoubleType(), True),                 # float64
    types.StructField('trip_distance', types.DoubleType(), True),                   # float64
    types.StructField('fare_amount', types.DoubleType(), True),                     # float64
    types.StructField('extra', types.DoubleType(), True),                           # float64
    types.StructField('mta_tax', types.DoubleType(), True),                         # float64
    types.StructField('tip_amount', types.DoubleType(), True),                      # float64
    types.StructField('tolls_amount', types.DoubleType(), True),                    # float64
    types.StructField('ehail_fee', types.DoubleType(), True),                        # object
    types.StructField('improvement_surcharge', types.DoubleType(), True),           # float64
    types.StructField('total_amount', types.DoubleType(), True),                    # float64
    types.StructField('payment_type', types.DoubleType(), True),                    # float64
    types.StructField('trip_type', types.DoubleType(), True),                       # float64
    types.StructField('congestion_surcharge', types.StringType(), True),            # float64
])

yellow_schema = types.StructType([
    types.StructField('VendorID', types.IntegerType(), True),       # int64                          
    types.StructField('tpep_pickup_datetime', types.TimestampType(), True),      # datetime64[us]     
    types.StructField('tpep_dropoff_datetime', types.TimestampType(), True),     # datetime64[us]    
    types.StructField('passenger_count', types.DoubleType(), True),     # float64                  
    types.StructField('trip_distance', types.DoubleType(), True),      # float64                      
    types.StructField('RatecodeID', types.DoubleType(), True),     # float64                      
    types.StructField('store_and_fwd_flag', types.StringType(), True),     # str                      
    types.StructField('PULocationID', types.IntegerType(), True),        # int64                 
    types.StructField('DOLocationID', types.IntegerType(), True),        # int64                   
    types.StructField('payment_type', types.IntegerType(), True),        # int64                     
    types.StructField('fare_amount', types.DoubleType(), True),     # float64                           
    types.StructField('extra', types.DoubleType(), True),       # float64                         
    types.StructField('mta_tax', types.DoubleType(), True),     # float64                      
    types.StructField('tip_amount', types.DoubleType(), True),      # float64                    
    types.StructField('tolls_amount', types.DoubleType(), True),        # float64                        
    types.StructField('improvement_surcharge', types.DoubleType(), True),       # float64           
    types.StructField('total_amount', types.DoubleType(), True),        # float64                    
    types.StructField('congestion_surcharge', types.DoubleType(), True),       # float64                    
    types.StructField('airport_fee', types.DoubleType(), True),        # object                       
])

```

```python
YEARS = [2020,2021]
TAXI_TYPES = ["green","yellow"]

for taxi_type in TAXI_TYPES:
    for year in YEARS:
        INPUT_PREFIX = f"../data/parquet/{taxi_type}/{year}"
        OUTPUT_PREFIX = f"../data/parquet_spark/{taxi_type}/{year}"

        for month in range(1, 13):
            MONTH = f"{month:02d}"
            INPUT_FILE  = f"{INPUT_PREFIX}/{MONTH}/{taxi_type}_tripdata_{year}-{MONTH}.parquet"
            OUTPUT_FILE = f"{OUTPUT_PREFIX}/{MONTH}/{taxi_type}_tripdata_{year}-{MONTH}.parquet"

            print(f"Processing {INPUT_FILE} -> {OUTPUT_FILE}")
            spark.read.parquet(INPUT_FILE) \
                .repartition(4) \
                .write \
                    .mode("overwrite") \
                    .parquet(OUTPUT_FILE)

            print(f"Processing Completed: {OUTPUT_FILE}")
```
